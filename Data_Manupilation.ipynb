```python
import pandas as pd
df = pd.read_csv ("ds.csv")
df.head()
```

Filtered Data Frames

```python


# Select only 'name' and 'city' columns
df2 = df[['Gender', 'Age']]
df2.head()
```

Conditional Filtering

```python
# Select rows where age > 30, and show only 'name' and 'age'
df3 = df[df['Age'] > 30][['Track', 'EmployeeNo','Age']]
df3.head()
```

Filter within a Column

```python
# Get rows where 'city' is 'NY'
df4 = df[df['Gender'] == 'Male']
df4.head()
```

```python
df5 = df[df['Track'].isin(['Technology', 'Qulaity'])]
df5.head()
```

```python
# Get rows where age > 25 and gender is 'Male'= String within ""
df6 = df[(df['Age'] > 40) & (df['Gender'] == 'Male')]
df6.head()
```

```python
# Get rows where age > 40 and Tier is '1' NO ""
df7 = df[(df['Age'] > 40) & (df['Tier'] == 1)]
df7.head()
```
